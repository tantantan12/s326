# Step 2: Adjacency Matrix
#we construct an adjacency matrix by binding the rows.
#Matrices are the R objects in which the elements 
#are arranged in a two-dimensional rectangular layout.
#rbind() function combines vector, matrix or data frame by rows.

```r

netmat1 <- rbind(c(0,1,1,0,0),  c(0,0,1,1,0),  +
                   c(0,1,0,0,0),  c(0,0,0,0,0),  c(0,0,1,0,0))

#define the row names and column names of the matrix.
rownames(netmat1) <- c("A","B","C","D","E") 
colnames(netmat1) <- c("A","B","C","D","E")  
```

# Step 3: Construct and Visualize the network

```r
#construct a network object based on the adjacency matrix.
net1 <- network(netmat1,matrix.type="adjacency") 

#plot this network
gplot(net1,   displaylabels = TRUE) 
```
